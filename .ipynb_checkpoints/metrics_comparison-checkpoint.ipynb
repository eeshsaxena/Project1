```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
import warnings
warnings.filterwarnings('ignore')

# ─── Colour palette ──────────────────────────────────────────────────────────
C_KG  = "#e05252"   # red   — KG-RAG
C_V4  = "#f0a832"   # amber — TruthfulRAG v4
C_V5  = "#3dbf7c"   # green — Enhanced v5

plt.rcParams.update({
    "figure.facecolor":  "#0f1117",
    "axes.facecolor":    "#1a1d27",
    "axes.edgecolor":    "#3a3d4d",
    "axes.labelcolor":   "#d0d3e0",
    "xtick.color":       "#9093a0",
    "ytick.color":       "#9093a0",
    "text.color":        "#d0d3e0",
    "grid.color":        "#2a2d3a",
    "grid.linestyle":    "--",
    "grid.alpha":        0.5,
    "font.family":       "DejaVu Sans",
    "legend.facecolor":  "#1a1d27",
    "legend.edgecolor":  "#3a3d4d",
})

print("Setup complete ✓")
```

```
# ─── Metric data ─────────────────────────────────────────────────────────────
metrics = [
    "Conflict\nDetection",
    "Retrieval\nRecall",
    "Historical\nAccuracy",
    "Adversarial\nRobustness",
    "Multi-Hop\nAccuracy",
    "Explainability",
    "Confidence\nOutput",
]

kg_rag = [40, 60, 30, 35, 25,  0,  0]
v4     = [87, 73, 54, 61, 48,  0,  0]
v5     = [96, 89, 91, 84, 72, 100, 100]

# Score-gap data for temporal conflict resolution
labels_gap = ["KG-RAG", "TruthfulRAG v4", "Enhanced v5"]
gaps       = [0.000,     0.013,            0.452]

# LLM samples per query
samples = [5, 3, "1-3\n(adaptive)"]

# Score formula components per system
components = {
    "KG-RAG":          [1, 0, 0, 0, 0],   # only Ref(p)
    "TruthfulRAG v4":  [1, 1, 1, 0, 0],   # + hub + PPR
    "Enhanced v5":     [1, 1, 1, 1, 1],   # + decay + corroborate
}
comp_labels = ["Ref(p)", "Hub Pen", "PPR", "Decay\n[N2]", "Corroborate\n[N1]"]

print("Data ready ✓")
```

```
fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

x   = np.arange(len(metrics))
w   = 0.26

b1 = ax.bar(x - w,   kg_rag, w, label="KG-RAG",          color=C_KG, alpha=0.9, zorder=3)
b2 = ax.bar(x,        v4,    w, label="TruthfulRAG v4",   color=C_V4, alpha=0.9, zorder=3)
b3 = ax.bar(x + w,    v5,    w, label="Enhanced v5",      color=C_V5, alpha=0.9, zorder=3)

for bars in [b1, b2, b3]:
    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 1.5,
                f"{h}%", ha="center", va="bottom", fontsize=7.5, color="#d0d3e0")

ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=9)
ax.set_ylabel("Score (%)", fontsize=11)
ax.set_ylim(0, 115)
ax.set_title("Pipeline Metric Comparison: KG-RAG vs TruthfulRAG v4 vs Enhanced v5",
             fontsize=13, fontweight="bold", color="white", pad=14)
ax.legend(fontsize=10, loc="upper left")
ax.yaxis.grid(True, zorder=0)
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("chart1_grouped_bar.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 1 saved ✓")
```

```
radar_labels = ["Conflict\nDetection", "Retrieval\nRecall",
                "Historical\nAccuracy", "Adversarial\nRobustness",
                "Multi-Hop\nAccuracy", "Explainability"]
N = len(radar_labels)
angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

def radar_row(vals): return vals + vals[:1]

kg_r = radar_row([v/100 for v in [40, 60, 30, 35, 25,  0]])
v4_r = radar_row([v/100 for v in [87, 73, 54, 61, 48,  0]])
v5_r = radar_row([v/100 for v in [96, 89, 91, 84, 72, 100]])

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

ax.plot(angles, kg_r, color=C_KG, linewidth=2, label="KG-RAG")
ax.fill(angles, kg_r, color=C_KG, alpha=0.15)
ax.plot(angles, v4_r, color=C_V4, linewidth=2, label="TruthfulRAG v4")
ax.fill(angles, v4_r, color=C_V4, alpha=0.15)
ax.plot(angles, v5_r, color=C_V5, linewidth=2, label="Enhanced v5")
ax.fill(angles, v5_r, color=C_V5, alpha=0.20)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(radar_labels, fontsize=9, color="#d0d3e0")
ax.set_yticks([0.25, 0.50, 0.75, 1.00])
ax.set_yticklabels(["25%","50%","75%","100%"], fontsize=7.5, color="#9093a0")
ax.tick_params(axis="y", colors="#2a2d3a")
ax.spines["polar"].set_color("#3a3d4d")
ax.grid(color="#2a2d3a", linestyle="--", alpha=0.6)

ax.set_title("Radar: Overall Pipeline Capability", fontsize=13,
             fontweight="bold", color="white", pad=18)
ax.legend(loc="upper right", bbox_to_anchor=(1.3, 1.1), fontsize=9)

plt.tight_layout()
plt.savefig("chart2_radar.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 2 saved ✓")
```

```
fig, ax = plt.subplots(figsize=(7, 5))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

colors = [C_KG, C_V4, C_V5]
bars   = ax.bar(labels_gap, gaps, color=colors, width=0.45, zorder=3, alpha=0.9)

for bar, val in zip(bars, gaps):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
            f"{val:.3f}", ha="center", va="bottom", fontsize=12,
            fontweight="bold", color="white")

ax.axhline(0.013, color=C_V4, linestyle=":", linewidth=1.2, alpha=0.7)
ax.axhline(0.452, color=C_V5, linestyle=":", linewidth=1.2, alpha=0.7)

ax.set_ylabel("Score Gap (Winner − Loser)", fontsize=11)
ax.set_title("Conflict Resolution Score Gap\n(Bigger = More Robust Decision)",
             fontsize=12, fontweight="bold", color="white", pad=12)
ax.set_ylim(0, 0.55)
ax.yaxis.grid(True, zorder=0)
ax.set_axisbelow(True)

ax.annotate("35× bigger gap\nthan v4!", xy=(2, 0.452), xytext=(1.55, 0.47),
            fontsize=9, color=C_V5, fontweight="bold",
            arrowprops=dict(arrowstyle="->", color=C_V5, lw=1.5))

plt.tight_layout()
plt.savefig("chart3_score_gap.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 3 saved ✓")
```

```
years = np.arange(1990, 2027)
age   = 2026 - years
decay = np.exp(-0.08 * age)

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

ax.plot(years, decay, color=C_V5, linewidth=2.5, zorder=3)
ax.fill_between(years, decay, alpha=0.15, color=C_V5)

key_years = {2014: ("Modi PM start", C_KG),
             2020: ("Mid-term",       C_V4),
             2024: ("Gandhi PM start",C_V5)}

for yr, (label, col) in key_years.items():
    d = np.exp(-0.08 * (2026 - yr))
    ax.scatter([yr], [d], color=col, zorder=5, s=80)
    ax.annotate(f"{label}\ndecay={d:.3f}", xy=(yr, d),
                xytext=(yr - 2, d + 0.08),
                fontsize=8.5, color=col, fontweight="bold",
                arrowprops=dict(arrowstyle="->", color=col, lw=1.2),
                ha="right")

ax.set_xlabel("Year of Fact", fontsize=11)
ax.set_ylabel("Temporal Decay Weight  [N2]", fontsize=11)
ax.set_title("N2 — Temporal Decay Curve  (λ = 0.08 per year)",
             fontsize=12, fontweight="bold", color="white", pad=12)
ax.set_ylim(0, 1.1)
ax.yaxis.grid(True, zorder=0)
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("chart4_temporal_decay.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 4 saved ✓")
```

```
fig, ax = plt.subplots(figsize=(8, 3.5))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

systems = list(components.keys())
matrix  = np.array([components[s] for s in systems], dtype=float)

im = ax.imshow(matrix, cmap="YlGn", vmin=0, vmax=1, aspect="auto")

for i, system in enumerate(systems):
    for j, comp in enumerate(comp_labels):
        val  = matrix[i, j]
        text = "✓ YES" if val else "✗ NO"
        col  = "#0f1117" if val else "#9093a0"
        ax.text(j, i, text, ha="center", va="center",
                fontsize=10, fontweight="bold", color=col)

ax.set_xticks(range(len(comp_labels)))
ax.set_xticklabels(comp_labels, fontsize=9)
ax.set_yticks(range(len(systems)))
ax.set_yticklabels(systems, fontsize=9)
ax.set_title("Score Formula Components per System",
             fontsize=12, fontweight="bold", color="white", pad=12)
plt.colorbar(im, ax=ax, label="Included", ticks=[0,1])

plt.tight_layout()
plt.savefig("chart5_formula_heatmap.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 5 saved ✓")
```

```
support_counts = np.arange(1, 11)
bonus = np.log(1 + support_counts) * 0.4

fig, ax = plt.subplots(figsize=(8, 4))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

ax.plot(support_counts, bonus, color=C_V5, marker="o", linewidth=2.5,
        markersize=7, zorder=3)
ax.fill_between(support_counts, bonus, alpha=0.15, color=C_V5)

for sc, b in zip(support_counts[:5], bonus[:5]):
    ax.annotate(f"+{b:.3f}", xy=(sc, b), xytext=(sc+0.1, b+0.015),
                fontsize=8, color=C_V5)

ax.set_xlabel("Number of Documents Supporting the Fact", fontsize=11)
ax.set_ylabel("Corroboration Bonus  [N1]", fontsize=11)
ax.set_title("N1 — Corroboration Score Bonus  (log scale growth)",
             fontsize=12, fontweight="bold", color="white", pad=12)
ax.yaxis.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("chart6_corroboration.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 6 saved ✓")
```

```
scenarios = [
    "Clear entropy\n3 docs\n2024 fact",
    "Clear entropy\n1 doc\n2024 fact",
    "Clear entropy\n1 doc\n2014 fact",
    "Ambiguous\n1 doc\n2019 fact",
    "No entropy\n1 doc\n1995 fact",
]

h_sig  = [0.80, 0.80, 0.80, 0.20, 0.00]
sup_s  = [0.77, 0.39, 0.39, 0.39, 0.39]
rec_s  = [0.90, 0.90, 0.55, 0.70, 0.13]

w_h, w_s, w_r = 0.40, 0.30, 0.30
H_part   = [w_h*h for h in h_sig]
S_part   = [w_s*s for s in sup_s]
R_part   = [w_r*r for r in rec_s]
total    = [h+s+r for h,s,r in zip(H_part, S_part, R_part)]

x = np.arange(len(scenarios))
fig, ax = plt.subplots(figsize=(11, 5))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

ax.bar(x, H_part, label="Entropy signal (40%)", color="#5b9bd5", zorder=3, alpha=0.9)
ax.bar(x, S_part, bottom=H_part, label="Support signal (30%)", color="#ed7d31", zorder=3, alpha=0.9)
ax.bar(x, R_part, bottom=[h+s for h,s in zip(H_part,S_part)],
       label="Recency signal (30%)", color=C_V5, zorder=3, alpha=0.9)

for i, tot in enumerate(total):
    ax.text(i, tot + 0.012, f"{tot*100:.0f}%", ha="center",
            fontsize=10, fontweight="bold", color="white")

ax.set_xticks(x)
ax.set_xticklabels(scenarios, fontsize=8)
ax.set_ylabel("Confidence Score  [N7]", fontsize=11)
ax.set_ylim(0, 1.15)
ax.set_title("N7 — Calibrated Confidence Score Breakdown by Scenario",
             fontsize=12, fontweight="bold", color="white", pad=12)
ax.legend(fontsize=9, loc="upper right")
ax.yaxis.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("chart7_confidence.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 7 saved ✓")
```

```
gain_labels = metrics
gains = [v - u for v, u in zip(v5, v4)]

fig, ax = plt.subplots(figsize=(11, 5))
fig.patch.set_facecolor("#0f1117")
ax.set_facecolor("#1a1d27")

bar_cols = [C_V5 if g > 0 else C_KG for g in gains]
bars = ax.bar(gain_labels, gains, color=bar_cols, zorder=3, alpha=0.9, width=0.5)

for bar, g in zip(bars, gains):
    ax.text(bar.get_x() + bar.get_width()/2,
            g + (1 if g >= 0 else -3),
            f"+{g}%" if g >= 0 else f"{g}%",
            ha="center", fontsize=10, fontweight="bold", color="white")

ax.axhline(0, color="#3a3d4d", linewidth=1)
ax.set_ylabel("Percentage Gain over TruthfulRAG v4", fontsize=11)
ax.set_title("Enhanced v5 Gains over TruthfulRAG v4 (per metric)",
             fontsize=12, fontweight="bold", color="white", pad=12)
ax.yaxis.grid(True, zorder=0); ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("chart8_gains.png", dpi=150, bbox_inches="tight", facecolor="#0f1117")
plt.show()
print("Chart 8 saved ✓")
```
